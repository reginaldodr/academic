Como exemplo, vamos plotar as curvas de nível do parabolóide $z=9x^+4y^2$, no quadrado $[-5,5]\times[-5,5]$.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    return 9*x**2+4*y**2
    
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, 10, cmap='Reds')
#plt.contourf(X, Y, Z, 10, cmap='Reds') #fill
plt.colorbar()

plt.show()
```
