Exemplo2.1:Sejam A=(1,1), B=(2,2), C=(-1,0) e D=(0,1). Determine as coordenadas dos vetores $\overrightarrow{AB}$    e    $\overrightarrow{CD}$ e diga se são iguais.

```python
import sympy as sp
sp.init_printing(use_latex='mathjax')

A=sp.Matrix([1,1])
B=sp.Matrix([2,2])
C=sp.Matrix([-1,0])
D=sp.Matrix([0,1])

AB=B-A
print('Vetor AB')
display(AB)
print('Vetor CD')
CD=D-C
display(CD)

print('Verificando se são iguais')
AB==CD
```

Exemplo 2.2: Determinando as coordenadas do ponto D tal que $\overrightarrow{AB}=\overrightarrow{CD}$.

```python
import sympy as sp
sp.init_printing(use_latex='mathjax')

x,y=sp.symbols('x y',real=True)

A=sp.Matrix([0,1])
B=sp.Matrix([1,-1/2])
C=sp.Matrix([1,0])
D=sp.Matrix([x,y])


AB=B-A
print('Vetor AB')
display(AB)


CD=D-C
print('Vetor CD')
display(CD)

print('Equação')
eq=sp.Eq(AB,CD)
display(eq)

sol=sp.solve(eq,(x,y))

print('Solução da equação')

display(sol)
```

Exemplo 2.3: Dados os vetores $\overrightarrow{u}=(1,-2)$, $\overrightarrow{v}=(-3,2)$, determine o vetor $\overrightarrow{w}=2 \overrightarrow{u}+3 \overrightarrow{v}$ e sua norma.

```python
import sympy as sp

u=sp.Matrix([1,-2])
v=sp.Matrix([-3,2])

print('O vetor buscado é:')
w=2*u+3*v
display(w)

print('norma do vetor w')
display(w.norm())
```
